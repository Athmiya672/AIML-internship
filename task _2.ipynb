```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import math
df = pd.read_csv('/content/Titanic-Dataset (1).csv')
print(df.describe())
print(df.describe(include=['object']))
```

```python
numeric_cols = df.select_dtypes(include='number').columns

df[numeric_cols].hist(figsize=(12, 8), bins=20)
plt.suptitle("Histograms of Numeric Features", fontsize=16)
plt.tight_layout()
plt.show()

```

```python
numeric_cols = df.select_dtypes(include='number').columns
num_plots = len(numeric_cols)
cols = 3
rows = math.ceil(num_plots / cols)
plt.figure(figsize=(5 * cols, 4 * rows))

for i, col in enumerate(numeric_cols):
    plt.subplot(rows, cols, i + 1)
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')

plt.tight_layout()
plt.show()


```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.pairplot(df[['Age', 'Fare', 'Pclass', 'Survived']], hue='Survived')
plt.suptitle("Pairplot of Selected Features", y=1.02)
plt.show()
```

```python
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival Count by Gender')
plt.show()
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survival Count by Passenger Class')
plt.show()
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.title("Age vs Fare Colored by Survival")
plt.show()
```

```python
print("""
Inferences:
- Females had a higher survival rate compared to males.
- Passengers in 1st class were more likely to survive than those in 3rd class.
- Younger passengers and those who paid higher fares showed higher survival probabilities.
- Some outliers in 'fare' and 'age' (e.g., very high fare values).
""")
```
